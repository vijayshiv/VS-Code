```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# this is the url of the website
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops/'
r = requests.get(url)

# collected the data in html format from the website using BeautifulSoup
soup = BeautifulSoup(r.text, 'lxml')
print(soup)
```

```python
# this the data from the ads box
box = soup.find_all('div', class_ = 'col-sm-4 col-lg-4 col-md-4')
print(box)
```

```python
# extracting name of the products from all the ads
name = soup.find_all('a', class_ = 'title')
# print(name)

# using list comprehension to get the list of all name of the products
names = [data.text for data in name]
print(names)
```

```python
# extracting prices in the same way
price = soup.find_all('h4', class_ = 'pull-right price')
# print(price)

prices = [data.text for data in price]
print(prices)
```

```python
# extracting description in the same way as well
description = soup.find_all('p', class_ = 'description')
# print(price)

descriptions = [data.text for data in description]
print(descriptions)
```

```python
# extracting reviews i.e. how many review we got for each product
review = soup.find_all('p', class_ = 'pull-right')
# print(review)

#getting the review number only
reviews = [data.text.split(' ')[0] for data in review]
print(reviews)
```

```python
# created the dataframe column header for the table

tabledata = pd.DataFrame(list(zip(names,prices,descriptions,reviews)),columns=['name_of_product', 'price_of_product','description', 'total_no_of_reviews'])
tabledata
```
